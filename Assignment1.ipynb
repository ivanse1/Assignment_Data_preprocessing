```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
chronic_kidney_disease = fetch_ucirepo(id=336) 
  
# data (as pandas dataframes) 
X = chronic_kidney_disease.data.features 
y = chronic_kidney_disease.data.targets 
```

```python
import pandas as pd

dfX = pd.DataFrame(X)
dfY = pd.DataFrame(y)
```

```python
dfX
```

```python
dfX = dfX.drop(columns=['rbc', 'pc', 'pcc', 'ba', 'sc', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane'])
```

```python
dfX
```

```python
dfY
```

```python
df = dfX.join(dfY, how='left')
```

```python
df
```

```python
df['hemo'] = df['hemo'] * 10
```

```python
df
```

```python
df['class'] = df['class'].replace({'ckd': 'a', 'ckd\t': 'a', 'notckd': 'c'})
```

```python
df
```

```python
# Remove rows with three or more missing values
df_cleaned = df.dropna(thresh=len(df.columns) - 2)
```

```python
remaining_rows = len(df_cleaned)
print(f"Remaining rows: {remaining_rows}")
```

```python
df_cleaned
```

```python
affected_df = df_cleaned[df_cleaned['class'] == 'a']
print(f"Affected rows: {len(affected_df)}")
affected_df
```

```python
control_df = df_cleaned[df_cleaned['class'] == 'c']
print(f"Control rows: {len(control_df)}")
control_df
```

```python
affected_stats = affected_df.describe()
print(affected_stats)
```

```python
control_stats = control_df.describe()
print(control_stats)
```

```python
import matplotlib.pyplot as plt

# Function to plot histograms for numerical columns
def plot_histograms(df, title_prefix):
    numerical_cols = df.select_dtypes(include=['number']).columns
    num_cols = len(numerical_cols)
    
    if num_cols == 0:
        print("No numerical columns to plot.")
        return
    
    plt.figure(figsize=(15, 5 * num_cols))
    
    for i, col in enumerate(numerical_cols):
        plt.subplot(num_cols, 1, i + 1)
        df[col].hist(bins=30, edgecolor='black')
        plt.title(f'{title_prefix} - {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
    
    plt.tight_layout()
    plt.show()

# Plot histograms for each DataFrame
plot_histograms(affected_df, 'Affected')
plot_histograms(control_df, 'Control')
```

```python
# Max value in Affected-bp is 180, which is an outlier, it should be removed
affected_df = affected_df[affected_df['bp'] < 180]
```

```python
# Max value in Affected-pop is over 40, which is an outlier, it should be removed
affected_df = affected_df[affected_df['pot'] < 40]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Function to plot the correlation matrix
def plot_correlation_matrix(df, title):
    # Select only numerical columns
    numerical_df = df.select_dtypes(include=['number'])
    
    if numerical_df.empty:
        print(f"No numerical columns in {title}")
        return
    
    # Calculate the correlation matrix
    corr_matrix = numerical_df.corr()
    
    # Plot the heatmap
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title(title)
    plt.show()

# Plot the correlation matrix for each DataFrame
plot_correlation_matrix(affected_df, 'Correlation Matrix for Affected Individuals')
plot_correlation_matrix(control_df, 'Correlation Matrix for Control Individuals')
```
